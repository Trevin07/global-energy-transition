```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fossil_fuel = pd.read_csv('fossil_fuel.csv')
co2_emissions = pd.read_csv('co2_emissions.csv')
renewable_energy = pd.read_csv('renewable_energy.csv',)
```

```python
fossil_fuel_melted = pd.melt(fossil_fuel, id_vars=["Location"], var_name="Year", value_name="Fossil Fuel Consumption")
co2_emissions_melted = pd.melt(co2_emissions, id_vars=["Location"], var_name="Year", value_name="CO2 Emissions")
renewable_energy_melted = pd.melt(renewable_energy, id_vars=["Location"], var_name="Year", value_name="Renewable Energy Consumption")


merged_data = fossil_fuel_melted.merge(co2_emissions_melted, on=["Location", "Year"])
merged_data = merged_data.merge(renewable_energy_melted, on=["Location", "Year"])

print(merged_data.head())
```

```python
print(merged_data.info())
```

```python

print(merged_data.describe())

print(merged_data.isnull().sum())
```

```python
top_countries = merged_data['Location'].unique()[:5]
top_countries_data = merged_data[merged_data['Location'].isin(top_countries)]
```

```python
#plot of fossil fuel consumption over time of the first 5 countres
plt.figure(figsize=(12, 6))
sns.lineplot(data=top_countries_data, x="Year", y="Fossil Fuel Consumption", hue="Location")
plt.title('Fossil Fuel Consumption Over Time')
plt.xlabel('Year')
plt.ylabel('Fossil Fuel Consumption')
plt.xticks(rotation=45)
plt.legend(title="Countries")
plt.show()
```

```python

correlation_data = merged_data[['Fossil Fuel Consumption', 'CO2 Emissions', 'Renewable Energy Consumption']]


corr_matrix = correlation_data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#C02 emissions  over the years of  group of countries 
import matplotlib.pyplot as plt
import seaborn as sns

total_emissions = merged_data.groupby("Location")["CO2 Emissions"].sum()
top10 = total_emissions.sort_values(ascending=False).head(7).index

filtered_data = merged_data[merged_data["Location"].isin(top10)]

plt.figure(figsize=(14, 7))
sns.lineplot(data=filtered_data, x="Year", y="CO2 Emissions", hue="Location")
plt.title('Carbon dioxide (CO2) emissions (total) excluding LULUCF (Mt CO2e)(Top 7 Categories)')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions')
plt.xticks(rotation=45)
plt.legend(title="Country Group")
plt.show()

```

```python
#average fossil fuel consumption per year across countries
yearly_avg_fossil_fuel = merged_data.groupby('Year')['Fossil Fuel Consumption'].mean()

plt.figure(figsize=(12, 6))
sns.lineplot(x=yearly_avg_fossil_fuel.index, y=yearly_avg_fossil_fuel.values)
plt.title('Average Fossil Fuel Consumption % Over Time')
plt.xlabel('Year')
plt.ylabel('Average Fossil Fuel Consumption')
plt.xticks(rotation=45)
plt.show()
```
